```python
df1 = pd.DataFrame({ #creating new dataframe
    "city":["new york", "delhi" , "mumbai" , "chicago"],
    "temperature":[21,55,48, 11],
})
df1
```

```python
df2 = pd.DataFrame({
    "city":["new york", "delhi", "mumbai", "chicago"],
    "humidity":[65,68,75,55]
})
df2
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="inner") # just like in databases there are joins - here the inner join will add only the common elements whcih is intersection
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer") # just like in databases there are joins - here the outer join will add all the elements which is union 
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left") # just like in databases there are joins - here the left join will take the common elements as well as the data from left table
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="right") # just like in databases there are joins - here the right join will take the common elements as well as the data from right table
df3
```
